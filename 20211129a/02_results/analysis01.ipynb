# Accuracy analysis

## The data

```python
import numpy as np

data = np.array([[12.0, 0.00696, 0.00005],
                [12.5, 0.00948, 0.00007],
                [13.0, 0.01313, 0.00014],
                [13.5, 0.01826, 0.00021],
                [14.0, 0.02690, 0.00033],
                [14.5, 0.03957, 0.00058]])
```

## Plotting the data.

```python
import matplotlib.pylab as plt

plt.errorbar(data[:,0], data[:,1], yerr=data[:,2], capsize=5, fmt='o', markersize=2, 
             ecolor='black', markeredgecolor = "black", color='w', label='Simulation')
plt.yscale('log')
plt.xlabel('Object magnitude (mag)')
plt.ylabel('Evaluated accuracy (px)')
plt.title('~400 Gaussian objects')
plt.legend()
plt.show()
```

```python
mag = np.arange(12.,15.,0.5)
phr = 10.**((mag-12)/(-2.5))

plt.errorbar(phr, data[:,1], yerr=data[:,2], capsize=5, fmt='o', markersize=2, 
             ecolor='black', markeredgecolor = "black", color='w', label='Simulation')
plt.plot(phr, 7.e-3*phr**-.75, label='$\propto$(photon flux)$^{-0.75}$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Photon flux ratio')
plt.ylabel('Evaluated accuracy (px)')
plt.title('~400 Gaussian objects')
plt.legend()
plt.show()
```
