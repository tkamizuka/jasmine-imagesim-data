```python
import astropy.io.fits as pf
import matplotlib.pylab as plt
from scipy.ndimage import shift
import numpy as np
```

```python
fits1 = pf.open("psf.fits")[0]
fits2 = pf.open("psf_wo_tilt.fits")[0]
```

```python
arr2 = []
diff_imgs = []
for i in range(0,400):
    print(i)
    arr=[]
    for dy in np.arange(-1.,1.1,0.1):
        for dx in np.arange(-1.,1.1,0.1):
            arr.append([dx,dy,np.sum(np.abs(fits2.data[i]-shift(fits1.data[i],(dy,dx))))])
    arr=np.array(arr)
    
    pos = np.argmin(arr[:,2])
    dxtmp, dytmp, sumdiff = arr[pos]

    arr = []
    for dy in np.arange(dytmp-0.09, dytmp+0.1, 0.01):
        for dx in np.arange(dxtmp-0.09, dxtmp+0.1, 0.01):
            arr.append([dx,dy,np.sum(np.abs(fits2.data[i]-shift(fits1.data[i],(dy,dx))))])
    arr=np.array(arr)
    
    pos = np.argmin(arr[:,2])
    dxtmp, dytmp, sumdiff = arr[pos]

    arr = []
    for dy in np.arange(dytmp-0.009, dytmp+0.01, 0.001):
        for dx in np.arange(dxtmp-0.009, dxtmp+0.01, 0.001):
            arr.append([dx,dy,np.sum(np.abs(fits2.data[i]-shift(fits1.data[i],(dy,dx))))])
    arr=np.array(arr)
    
    pos = np.argmin(arr[:,2])
    dxtmp, dytmp, sumdiff = arr[pos]
    
    arr = []
    for dy in np.arange(dytmp-0.0009, dytmp+0.001, 0.0001):
        for dx in np.arange(dxtmp-0.0009, dxtmp+0.001, 0.0001):
            arr.append([dx,dy,np.sum(np.abs(fits2.data[i]-shift(fits1.data[i],(dy,dx))))])
    arr=np.array(arr)
    
    pos = np.argmin(arr[:,2])
    dx, dy, sumdiff = arr[pos]
    shifted = shift(fits1.data[i], (dy, dx))
    arr2.append(arr[pos])
    diff_imgs.append(fits2.data[i]-shifted)
arr2 = np.array(arr2)
diff_imgs = np.array(diff_imgs)
```
