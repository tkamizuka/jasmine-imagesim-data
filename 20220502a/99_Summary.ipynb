# Summary of the results

We summarize the results stored in the directories 02_results*.

## Importing some modules.

```python
import numpy as np
import matplotlib.pylab as plt
```

## Loading data.

```python
data = {}
data['ctl0_det00_tel00'] = np.loadtxt('02_results_ctl0_det00_tel00/data_accuracy.dat')
data['ctl1_det00_tel00'] = np.loadtxt('02_results_ctl1_det00_tel00/data_accuracy.dat')
data['ctl2_det00_tel00'] = np.loadtxt('02_results_ctl2_det00_tel00/data_accuracy.dat')
data['ctl0_det00_tel02'] = np.loadtxt('02_results_ctl0_det00_tel02/data_accuracy.dat')
data['ctl1_det00_tel02'] = np.loadtxt('02_results_ctl1_det00_tel02/data_accuracy.dat')
data['ctl2_det00_tel02'] = np.loadtxt('02_results_ctl2_det00_tel02/data_accuracy.dat')
data['ctl2_det10_tel10'] = np.loadtxt('02_results_ctl2_det10_tel10/data_accuracy.dat')
data['ctl2_det10_tel12'] = np.loadtxt('02_results_ctl2_det10_tel12/data_accuracy.dat')
```

## Comparison (Det00/Tel00)

```python
ymin_px = 3e-3
ymax_px = 2e-2
efl = 3892.308e-3 # m
pix_size = 10e-6  # m
fct = pix_size/efl*180.*3600./np.pi

fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.errorbar(data['ctl0_det00_tel00'][:,0], data['ctl0_det00_tel00'][:,1], yerr=data['ctl0_det00_tel00'][:,2],\
            marker='o', linestyle='None', label='RealPSF')
ax.errorbar(data['ctl1_det00_tel00'][:,0], data['ctl1_det00_tel00'][:,1], yerr=data['ctl1_det00_tel00'][:,2],\
            marker='x', linestyle='None', label='RealPSF+GaussACE')
ax.errorbar(data['ctl2_det00_tel00'][:,0], data['ctl2_det00_tel00'][:,1], yerr=data['ctl2_det00_tel00'][:,2],\
            marker='s', linestyle='None', label='RealPSF+GaussACE+Flat')
ax.set_yscale('log')
ax2.set_yscale('log')
ax.legend()
ax.set_xlabel('Object magnitude (mag)')
ax.set_ylabel('Position accuracy (px)')
ax2.set_ylabel('Position accuracy (arcsec)')
ax.set_title('Det00/Tel00')
ax.set_ylim([ymin_px, ymax_px])
ax2.set_ylim([ymin_px*fct, ymax_px*fct])
plt.show()
```

## Comparison (Det00/Tel02)

```python
ymin_px = 3e-3
ymax_px = 2e-2
efl = 4369.231e-3 # m
pix_size = 10e-6  # m
fct = pix_size/efl*180.*3600./np.pi

fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.errorbar(data['ctl0_det00_tel02'][:,0], data['ctl0_det00_tel02'][:,1], yerr=data['ctl0_det00_tel02'][:,2],\
            marker='o', linestyle='None', label='RealPSF')
ax.errorbar(data['ctl1_det00_tel02'][:,0], data['ctl1_det00_tel02'][:,1], yerr=data['ctl1_det00_tel02'][:,2],\
            marker='x', linestyle='None', label='RealPSF+GaussACE')
ax.errorbar(data['ctl2_det00_tel02'][:,0], data['ctl2_det00_tel02'][:,1], yerr=data['ctl2_det00_tel02'][:,2],\
            marker='s', linestyle='None', label='RealPSF+GaussACE+Flat')
ax.set_yscale('log')
ax2.set_yscale('log')
ax.legend()
ax.set_xlabel('Object magnitude (mag)')
ax.set_ylabel('Position accuracy (px)')
ax2.set_ylabel('Position accuracy (arcsec)')
ax.set_title('Det00/Tel02')
ax.set_ylim([ymin_px, ymax_px])
ax2.set_ylim([ymin_px*fct, ymax_px*fct])
plt.show()
```

## Comparison (Ctl2)

```python
efl1 = 3892.308e-3 # m
efl2 = 4369.231e-3 # m
efl3 = 4107.692e-3 # m
efl4 = 4630.769e-3 # m
pix_size = 10e-6   # m
fct1 = pix_size/efl1*180.*3600./np.pi
fct2 = pix_size/efl2*180.*3600./np.pi
fct3 = pix_size/efl3*180.*3600./np.pi
fct4 = pix_size/efl4*180.*3600./np.pi

fig, ax = plt.subplots()
ax.errorbar(data['ctl2_det00_tel00'][:,0], data['ctl2_det00_tel00'][:,1]*fct1, yerr=data['ctl2_det00_tel00'][:,2]*fct1,\
            marker='o', linestyle='None', label='EPD:340; EFL:3892.308')
ax.errorbar(data['ctl2_det00_tel02'][:,0], data['ctl2_det00_tel02'][:,1]*fct2, yerr=data['ctl2_det00_tel02'][:,2]*fct2,\
            marker='x', linestyle='None', label='EPD:340; EFL:4369.231')
ax.errorbar(data['ctl2_det10_tel10'][:,0], data['ctl2_det10_tel10'][:,1]*fct3, yerr=data['ctl2_det10_tel10'][:,2]*fct3,\
            marker='s', linestyle='None', label='EPD:360; EFL:4107.692')
ax.errorbar(data['ctl2_det10_tel12'][:,0], data['ctl2_det10_tel12'][:,1]*fct4, yerr=data['ctl2_det10_tel12'][:,2]*fct4,\
            marker='*', linestyle='None', label='EPD:360; EFL:4630.769')
ax.set_yscale('log')
ax.legend()
ax.set_xlabel('Object magnitude (mag)')
ax.set_ylabel('Position accuracy (arcsec)')
ax.set_title('RealPSF+GaussACE+Flat')
ax.set_ylim([2e-3, 1e-2])
plt.show()
```
