```python
import numpy as np
import matplotlib.pylab as plt
```

```python
data = []
data.append(np.loadtxt('02_results_ctl0_det00_tel00/data_accuracy.dat'))
data.append(np.loadtxt('02_results_ctl1_det00_tel00/data_accuracy.dat'))
data.append(np.loadtxt('02_results_ctl2_det00_tel00/data_accuracy.dat'))
data = np.array(data)
```

```python
ymin_px = 3e-3
ymax_px = 2e-2
efl = 4107.692e-3 # m
pix_size = 10e-6  # m
fct = pix_size/efl*180.*3600./np.pi

fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.errorbar(data[0,:,0], data[0,:,1], yerr=data[0,:,2], marker='.', linestyle='None', label='RealPSF')
ax.errorbar(data[1,:,0], data[1,:,1], yerr=data[1,:,2], marker='x', linestyle='None', label='RealPSF+GaussACE')
ax.errorbar(data[2,:,0], data[2,:,1], yerr=data[2,:,2], marker='*', linestyle='None', label='RealPSF+GaussACE+Flat')
ax.set_yscale('log')
ax2.set_yscale('log')
ax.legend()
ax.set_xlabel('Object magnitude (mag)')
ax.set_ylabel('Position accuracy (px)')
ax2.set_ylabel('Position accuracy (arcmin)')
ax.set_title('Det00/Tel00')
ax.set_ylim([ymin_px, ymax_px])
ax2.set_ylim([ymin_px*fct, ymax_px*fct])
plt.show()
```
