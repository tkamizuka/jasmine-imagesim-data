```python
import numpy as np
import matplotlib.pylab as plt
```

```python
data = []
data.append(np.loadtxt('02_results_ctl0_det00_tel00/data_accuracy.dat'))
data.append(np.loadtxt('02_results_ctl1_det00_tel00/data_accuracy.dat'))
data.append(np.loadtxt('02_results_ctl2_det00_tel00/data_accuracy.dat'))
data.append(np.loadtxt('02_results_ctl2_det00_tel02/data_accuracy.dat'))
data.append(np.loadtxt('02_results_ctl2_det10_tel10/data_accuracy.dat'))
data = np.array(data)
```

```python
ymin_px = 3e-3
ymax_px = 2e-2
efl = 3892.308e-3 # m
pix_size = 10e-6  # m
fct = pix_size/efl*180.*3600./np.pi

fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.errorbar(data[0,:,0], data[0,:,1], yerr=data[0,:,2], marker='.', linestyle='None', label='RealPSF')
ax.errorbar(data[1,:,0], data[1,:,1], yerr=data[1,:,2], marker='x', linestyle='None', label='RealPSF+GaussACE')
ax.errorbar(data[2,:,0], data[2,:,1], yerr=data[2,:,2], marker='*', linestyle='None', label='RealPSF+GaussACE+Flat')
ax.set_yscale('log')
ax2.set_yscale('log')
ax.legend()
ax.set_xlabel('Object magnitude (mag)')
ax.set_ylabel('Position accuracy (px)')
ax2.set_ylabel('Position accuracy (arcmin)')
ax.set_title('Det00/Tel00')
ax.set_ylim([ymin_px, ymax_px])
ax2.set_ylim([ymin_px*fct, ymax_px*fct])
plt.show()
```

```python
ymin_px = 3e-3
ymax_px = 2e-2
efl1 = 3892.308e-3 # m
efl2 = 4369.231e-3 # m
efl3 = 4107.692e-3 # m
pix_size = 10e-6   # m
fct1 = pix_size/efl1*180.*3600./np.pi
fct2 = pix_size/efl2*180.*3600./np.pi
fct3 = pix_size/efl3*180.*3600./np.pi

fig, ax = plt.subplots()
ax.errorbar(data[2,:,0], data[2,:,1]*fct1, yerr=data[2,:,2]*fct1, marker='*', linestyle='None', label='EPD:340; EFL:3892.308')
ax.errorbar(data[3,:,0], data[3,:,1]*fct2, yerr=data[3,:,2]*fct2, marker='*', linestyle='None', label='EPD:340; EFL:4369.231')
ax.errorbar(data[4,:,0], data[4,:,1]*fct3, yerr=data[4,:,2]*fct3, marker='*', linestyle='None', label='EPD:360; EFL:4107.692')
ax.set_yscale('log')
ax.legend()
ax.set_xlabel('Object magnitude (mag)')
ax.set_ylabel('Position accuracy (arcmin)')
ax.set_title('RealPSF+GaussACE+Flat')
ax.set_ylim([1e-3, 1e-2])
plt.show()
```
