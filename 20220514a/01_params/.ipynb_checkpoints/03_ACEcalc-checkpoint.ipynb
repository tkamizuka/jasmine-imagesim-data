# Examining ACE data

## Import modules

```python
import numpy as np
import matplotlib.pylab as plt
from jis.photonsim.ace import calc_ace
from jis.binutils.runphotonsim import run_calc_ace
from jis.photonsim.extract_json import ControlParams, Detector, AcePsd
```

## Load the settings

```python
cp = ControlParams.from_json("ctl4.json")
det = Detector.from_json("det00.json")
ace_psd = AcePsd.from_json("ace_001.json")
```

```python
for k in cp.ace_control.keys():
    print("{}: {}".format(k, cp.ace_control[k]))
```

## Calculate ACE

```python
acex, acey, Nts_per_plate = run_calc_ace(cp, det, ace_psd.parameters)
```

## Check the ACE data

```python
plt.plot(acex)
plt.plot(acey)
plt.show()
```

```python
print(acex.mean())
print(acex.std())
print(acey.mean())
print(acey.std())
r = np.sqrt(acex**2.+acey**2.)
print(np.std(r))
```

## Divide data into exposure sets

```python
texp=12.5
ndata = int(texp/cp.ace_control['dtace']+0.5)
nset = int(np.size(acex)/ndata)
print(ndata)
print(nset)

acex_d = np.reshape(acex[0:ndata*nset], (nset, ndata))
acey_d = np.reshape(acey[0:ndata*nset], (nset, ndata))
```

## Check the datasets

```python
for i in range(3):
    plt.plot(acex_d[i,:])
```

```python
for i in range(3):
    plt.plot(acey_d[i,:])
```

```python
plt.plot(np.std(acex_d, axis=1)/np.std(acex_d))
plt.plot(np.std(acey_d, axis=1)/np.std(acey_d))
```

```python
res = plt.hist(np.std(acex_d, axis=1))
```

```python
res = plt.hist(np.std(acey_d, axis=1))
```

```python
print(np.mean(np.std(acex_d, axis=1)))
print(np.mean(np.std(acey_d, axis=1)))
```

The standard deviation of the ace in each exposure is 0.87 times of that in 3600 sec. For setting the ace std in each exposure to be 300 mas, we have to set the ace std in 3600 sec to be 345 mas.
