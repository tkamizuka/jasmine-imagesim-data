# Study about the gaussian distribution

```python
from scipy.ndimage import gaussian_filter
import numpy as np
```

```python
a = np.zeros([101,101])
```

```python
a[50, 50]=1
```

```python
b=gaussian_filter(a, 2)
```

```python
import matplotlib.pylab as plt
```

```python
plt.imshow(b)
```

```python
plt.plot(np.sum(b, axis=0), '.-')
plt.plot(np.sum(b, axis=1), '.-')
```

```python
ind=np.indices((101,101))
```

```python
rs=np.sqrt((ind[0]-50.)**2.+(ind[1]-50.)**2.).flatten()
```

```python
plt.plot(rs,b.flatten())
plt.xlim([0,10])
plt.plot(rs,0.040*np.exp(-(rs-0.)**2./2./4.))
```

```python
x = np.random.normal(0,5,100000)
y = np.random.normal(0,5,100000)
print(np.std(x))
print(np.std(y))
print(np.std(np.sqrt(x**2.+y**2.)))
```

```python
plt.hist2d(x, y, bins=31)
plt.xlim([-20,20])
plt.ylim([-20,20])
plt.axes().set_aspect('equal')
```
